```python


!pip install transformers

from transformers import AutoModelForCausalLM, AutoTokenizer

model_id = 'stabilityai/stablelm-2-zephyr-1_6b'

tokenizer = AutoTokenizer.from_pretrained(model_id)
model = AutoModelForCausalLM.from_pretrained(model_id)

def chat_function(prompt_content):
    prompt = [{'role': 'user', 'content': prompt_content}]
    inputs = tokenizer.apply_chat_template(
        prompt,
        add_generation_prompt=True,
        return_tensors='pt'
    )

    tokens = model.generate(
        inputs.to(model.device),
        max_new_tokens=1024,
        temperature=0.5,
        do_sample=True
    )

    return tokenizer.decode(tokens[0], skip_special_tokens=False)


```

```python
import ipywidgets as widgets
from IPython.display import display, clear_output

# Define text input widget
text_input = widgets.Text(description='Input text:', placeholder='Enter text here')

# Define button widgets
button_submit = widgets.Button(description='Submit')
button_clear = widgets.Button(description='Clear')

# Define function to handle submit button click
def on_submit_button_click(b):
    print("User:", text_input.value)
    print("AI:", chat_function(text_input.value))

# Define function to handle clear button click
def on_clear_button_click(b):
    clear_output()

# Attach functions to button click events
button_submit.on_click(on_submit_button_click)
button_clear.on_click(on_clear_button_click)

# Display widgets
display(text_input, widgets.HBox([button_submit, button_clear]))


```
